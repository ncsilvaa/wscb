```python
from sklearn.model_selection import train_test_split
import numpy as np

df_test = []
def split(df: pd.DataFrame, train_size:float = 0.8):
    ratings_train = int(len(df) * train_size)  
    train, test = df.iloc[:ratings_train], df.iloc[ratings_train:]
    df_test.append(test)
    display(df), display(train), display(test)
    input("")
    return train

df_train = df_ratings.groupby("user_id").apply(lambda x: split(x.sort_values(["timestamp"]).reset_index(drop=True), train_size=0.8)).reset_index(drop=True)
df_test = pd.concat(df_test)
```
